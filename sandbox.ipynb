```python
import pandas as pd
import numpy as np
```

```python
data_path = "/home/kenny/osda/osda_datasets/raw/archive/heart.csv"
df = pd.read_csv(data_path)
```

```python
df
```

```python
col_replace = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[col_replace] = df[col_replace].replace(0, np.nan) # I will first replace 0 values with Nulls then calculate median.
med = df[col_replace].median()
df[col_replace] = df[col_replace].fillna(med)
```

```python
df.isnull().sum()
```

```python
df.to_csv("/home/kenny/osda/osda_datasets/processed/Healthcare-Diabetes.csv", index=False)
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv("/home/kenny/osda/osda_datasets/processed/heart.csv", index=False)
```

```python
# missing values filled with average
df["ph"].fillna(value = df["ph"].mean(), inplace=True)
df["Sulfate"].fillna(value = df["Sulfate"].mean(), inplace=True)
df["Trihalomethanes"].fillna(value = df["Trihalomethanes"].mean(), inplace=True)
df.isnull().sum()
```
